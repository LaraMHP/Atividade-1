Desafio 1-2: O pixel [0,0] é o primeiro pixel no canto superior esquerdo da imagem. Ele contém 3 valores (B, G, R) que representam as intensidades de azul, verde e vermelho, respectivamente.


```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("Imagem_VC.jpg")

# Converter de BGR para RGB para exibir corretamente
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Coordenadas do pixel [0,0]
x, y = 0, 0

# Desenhar um círculo no pixel [0,0] para destacá-lo
marcada = imagem_rgb.copy()
cv2.circle(marcada, (x, y), radius=10, color=(0, 0, 255), thickness=-1)  # Azul puro

# Mostrar imagem original e marcada lado a lado
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Imagem Original")
plt.imshow(imagem_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Pixel [0,0] marcado")
plt.imshow(marcada)
plt.axis("off")

plt.show()
```

Desafio 1-3


```python
import cv2

imagem = cv2.imread("Imagem_VC.jpg")
altura, largura, canais = imagem.shape

cantos = {
    "Superior esquerdo": (0, 0),
    "Superior direito": (0, largura-1),
    "Inferior esquerdo": (altura-1, 0),
    "Inferior direito": (altura-1, largura-1)
}

for nome, (y, x) in cantos.items():
    (b, g, r) = imagem[y, x]
    print(f"{nome} -> Vermelho: {r}, Verde: {g}, Azul: {b}")
```

Desafio 1-4


```python
import numpy as np

media_b = np.mean(imagem[:, :, 0])
media_g = np.mean(imagem[:, :, 1])
media_r = np.mean(imagem[:, :, 2])

print(f"Cor média -> Vermelho: {media_r:.2f}, Verde: {media_g:.2f}, Azul: {media_b:.2f}")
```

Desafio 1-5

```python
if media_r > media_g and media_r > media_b:
    print("A imagem tem predominância de VERMELHO.")
elif media_g > media_r and media_g > media_b:
    print("A imagem tem predominância de VERDE.")
else:
    print("A imagem tem predominância de AZUL.")
```

```python
import cv2
from matplotlib import pyplot as plt

# 1) Ler a imagem
imagem = cv2.imread("Imagem_VC.jpg")

# 2) Alterar um pixel específico
imagem[100, 100] =(0, 0, 255) # Azul puro em (100,100)

# 3) Salvar a imagem modificada
cv2.imwrite("Imagem_VC_modificada.jpg", imagem)

# 4) Exibir imagens lado a lado
original = cv2.imread("Imagem_VC.jpg")
modificada = cv2.imread("Imagem_VC_modificada.jpg")

# Converter para RGB para exibição correta
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
modificada = cv2.cvtColor(modificada, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Imagem Original")
plt.imshow(original)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Imagem Modificada")
plt.imshow(modificada)
plt.axis("off")

plt.show()
```
